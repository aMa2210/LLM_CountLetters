```python
import csv
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Freq spanish -> https://www.sttmedia.com/characterfrequency-spanish#letters (deleted those not selected)
# Freq english -> https://norvig.com/mayzner.html

df = pd.read_csv('Letter_Accuracy_Spa.csv')

df = df.sort_values(by='Freq', ascending=False)

df
columns = df.columns[2:].copy()
```

```python
for column in columns:
    df[f'err_{column}'] = 100 - df[column]
    if ('ori' in column):
        df.plot.bar(x='Letter', y=['Freq', f'err_{column}'], title=f"Error rate: {column}")
        plt.xticks(rotation=360)
        plt.ylabel("Adjusted Values")
        plt.ylim(0, 50)
        plt.show()

for column in columns:
    df[f'err_{column}'] = 100 - df[column]
    if (not 'ori' in column):
        df.plot.bar(x='Letter', y=['Freq', f'err_{column}'], title=f"Error rate: {column}")
        plt.xticks(rotation=360)
        plt.ylabel("Adjusted Values")
        plt.ylim(0, 50)
        plt.show()
```
